# Data processig and splitting in to three supdata train , val and test data






```python
import os
import random
import shutil

# Define the path to your data directory
data_dir = '/content/data'

# Define the output directory to save the train, val, and test sets
output_dir = '/content/data'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define the class names
class_names = ["covid", "normal", "lung_opacity" , "viral_pneumonia"]
# Define the train, val, and test ratios (e.g., 70%, 15%, 15%)
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Loop over each class folder
for class_name in class_names:
    # Create the train, val, and test subdirectories for each class
    train_dir = os.path.join(output_dir, 'train', class_name)
    val_dir = os.path.join(output_dir, 'val', class_name)
    test_dir = os.path.join(output_dir, 'test', class_name)

    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Get the list of image files for the current class
    class_dir = os.path.join(data_dir, class_name)
    image_files = os.listdir(class_dir)

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Calculate the number of samples for train, val, and test sets
    num_samples = len(image_files)
    num_train = int(train_ratio * num_samples)
    num_val = int(val_ratio * num_samples)

    # Split the image files into train, val, and test sets
    train_files = image_files[:num_train]
    val_files = image_files[num_train:num_train+num_val]
    test_files = image_files[num_train+num_val:]

    # Copy the image files to the respective train, val, and test directories
    for train_file in train_files:
        src = os.path.join(class_dir, train_file)
        dst = os.path.join(train_dir, train_file)
        shutil.copyfile(src, dst)

    for val_file in val_files:
        src = os.path.join(class_dir, val_file)
        dst = os.path.join(val_dir, val_file)
        shutil.copyfile(src, dst)

    for test_file in test_files:
        src = os.path.join(class_dir, test_file)
        dst = os.path.join(test_dir, test_file)
        shutil.copyfile(src, dst)
```

# Covid-19 classification using deep learning pretraind models CNN (vgg16, densenet121, resnet50 and resnet152).

```python
# Set random seed for reproducibility
torch.manual_seed(42)

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Define data transformations
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.25),
        transforms.RandomVerticalFlip(p=0.25),
        transforms.RandomRotation(degrees=15),
        transforms.ColorJitter(0.1, 0.1, 0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}
```

```python
# Load dataset
data_dir = '/content/data'
image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), data_transforms[x]) for x in ['train', 'val', 'test']}
```

```python
# Split dataset into train, validation, and test sets
train_data = image_datasets['train']
val_data = image_datasets['val']
test_data = image_datasets['test']
```

```python
# Print number of images after augmentation
print(f'Train Images: {len(train_data)}')
print(f'Validation Images: {len(val_data)}')
print(f'Test Images: {len(test_data)}')
```

```python
# Calculate class weights for data imbalance
train_class_sample_counts = [len(train_data[x]) for x in range(len(train_data.classes))]
val_class_sample_counts = [len(val_data[x]) for x in range(len(val_data.classes))]
test_class_sample_counts = [len(test_data[x]) for x in range(len(test_data.classes))]

train_class_weights = 1.0 / torch.tensor(train_class_sample_counts, dtype=torch.float32)
val_class_weights = 1.0 / torch.tensor(val_class_sample_counts, dtype=torch.float32)
test_class_weights = 1.0 / torch.tensor(test_class_sample_counts, dtype=torch.float32)

train_sample_weights = [train_class_weights[class_id] for class_id in train_data.targets]
val_sample_weights = [val_class_weights[class_id] for class_id in val_data.targets]
test_sample_weights = [test_class_weights[class_id] for class_id in test_data.targets]

train_sampler = WeightedRandomSampler(train_sample_weights, len(train_sample_weights), replacement=True)
val_sampler = WeightedRandomSampler(val_sample_weights, len(val_sample_weights), replacement=True)
test_sampler = WeightedRandomSampler(test_sample_weights, len(test_sample_weights), replacement=True)
```

```python
# Create data loaders
train_loader = DataLoader(train_data, batch_size=32, sampler=train_sampler)
val_loader = DataLoader(val_data, batch_size=32, sampler=val_sampler)
test_loader = DataLoader(test_data, batch_size=32, sampler=test_sampler)
```

```python
# Define the models
vgg16 = models.vgg16(pretrained=True)
densenet121 = models.densenet121(pretrained=True)
resnet50 = models.resnet50(pretrained=True)
resnet152 = models.resnet152(pretrained=True)
```

```python
# Replace the last fully connected layer for COVID-19 classification
num_classes = 4  # Number of classes: COVID-19, Normal, Other infections

vgg16.classifier[6] = nn.Linear(4096, num_classes)
densenet121.classifier = nn.Linear(1024, num_classes)
resnet50.fc = nn.Linear(2048, num_classes)
resnet152.fc = nn.Linear(2048, num_classes)
```

```python

# Define the loss function
criterion = nn.CrossEntropyLoss()
```

```python
# Define the optimizer
optimizer_vgg16 = optim.Adam(vgg16.parameters(), lr=0.0001)
optimizer_densenet121 = optim.Adam(densenet121.parameters(), lr=0.0001)
optimizer_resnet50 = optim.Adam(resnet50.parameters(), lr=0.0001)
optimizer_resnet152 = optim.Adam(resnet152.parameters(), lr=0.0001)
```

```python
# define the train function

import os

def train(model, optimizer, criterion, train_loader, val_loader, epochs=13, save_parent_dir='/content/drive/MyDrive/four_classes_weights'):
    model.to(device)

    train_losses = []
    train_accuracies = []
    val_losses = []
    val_accuracies = []

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss /= len(train_loader.dataset)
        train_acc = 100 * correct / total
        train_losses.append(train_loss)
        train_accuracies.append(train_acc)

        # Validation
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * images.size(0)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_loss /= len(val_loader.dataset)
        val_acc = 100 * correct / total
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

        print(f'Epoch: {epoch+1}/{epochs}, '
              f'Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.2f}%, '
              f'Val Loss: {val_loss:.4f}, Val Accuracy: {val_acc:.2f}%')

        # Save model weights
        model_name = model.__class__.__name__
        save_dir = os.path.join(save_parent_dir, model_name)
        os.makedirs(save_dir, exist_ok=True)
        save_path = os.path.join(save_dir, f'model_weights_{epoch+1}.pt')
        torch.save(model.state_dict(), save_path)
        print(f'Saved model weights to: {save_path}')

    return train_losses, train_accuracies, val_losses, val_accuracies


```

```python
import matplotlib.pyplot as plt

def plot_results(train_loss, val_loss, train_acc, val_acc):
    epochs = range(1, len(train_loss) + 1)

    plt.figure(figsize=(12, 4))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_loss, label='Train')
    plt.plot(epochs, val_loss, label='Validation')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_acc, label='Train')
    plt.plot(epochs, val_acc, label='Validation')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()



```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
#from fastai.basic_data import DataBunch
#from fastai.metrics import accuracy

def test(model, test_loader, model_name):
    model.to(device)
    model.eval()
    test_loss = 0.0
    correct = 0
    total = 0
    predictions = []
    true_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            predictions.extend(predicted.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    test_loss /= len(test_loader.dataset)
    accuracy = 100 * correct / total

    print(f'{model_name} - Test Loss: {test_loss:.4f}, Test Accuracy: {accuracy:.2f}%')

    # Create confusion matrix
    confusion_mat = confusion_matrix(true_labels, predictions)

    return confusion_mat
```

```python
# Visualize confusion matrix
def plot_confusion_matrix(confusion_mat, classes):
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion_mat, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
def visualize_accuracy(y_pred, classes):
    predictions = np.argmax(y_pred, axis=1)
    num_classes = len(classes)
    class_accuracy = np.zeros(num_classes)
    for i in range(num_classes):
        class_indices = np.where(predictions == i)[0]
        class_accuracy[i] = np.mean(predictions[class_indices] == i)
    plt.figure(figsize=(8, 6))
    plt.bar(classes, class_accuracy)
    plt.xlabel("Classes")
    plt.ylabel("Accuracy")
    plt.title("Accuracy per Class")
    plt.show()
```
